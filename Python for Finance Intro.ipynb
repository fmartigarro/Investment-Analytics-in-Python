```python
#initializing 'web' as the data attained from pandas' data reader package
from pandas_datareader import data as web
```

```python
#initializing 'Apple' data frame with the financial data attained from yahoo finance
Apple = web.DataReader('AAPL', data_source = 'yahoo', start = '1995-1-1')
```

```python
Apple
```

```python
#Apple.info() tells me information about the data
#"DataFrame" indicates its' data structure
#"6284" across all columns indicates theres no missing data

Apple.info()
```

```python
#Creating DF containing "Adjusted Close" column data from 5 different companies: Apple, P&G, Microsoft, Ford, GE

tickers = ['AAPL', 'PG', 'MSFT', 'F', 'GE']
new_data = pd.DataFrame()
for t in tickers:
    new_data[t] = web.DataReader(t, data_source = 'yahoo', start = '1995-1-1')['Adj Close']
```

```python
#printing the DF containing the "Adjusted Close" prices
new_data
```
